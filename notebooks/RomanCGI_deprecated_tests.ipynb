## Test the 721 mode.

```python
reload(wfirst)
spc721 = wfirst.CGI(mode='DISKSPC_F721')
spc721.options['source_offset_r'] = 0 # arcsec
spc721.options['source_offset_theta'] = 0 # deg w.r.t. North

plt.figure(figsize=(9,16))
spc721.display()
```

```python
lamc = 721e-9*u.m
D = 2.3631*u.m
lamD825 = lamc/D
as_per_lamD = lamc * 360.0 * 3600.0 / (2 * np.pi * D)    # as per lambda0/D
fov_as = 45*as_per_lamD.value

plt.figure(figsize=[12,9])
monopsf_onax = spc721.calc_psf(nlambda=1, fov_arcsec=fov_as, display=True)
```

## Calculate off-axis PSF.

```python
spc721.options['source_offset_r'] = 0.6 #10*as_per_lamD.value # arcsec
spc721.options['source_offset_theta'] = -45. # deg w.r.t. North

plt.figure(figsize=[12,9])
monopsf_offax = spc721.calc_psf(nlambda=1, fov_arcsec=fov_as, display=True)
```

## Display both PSFs side-by-side.

```python
fig, ax  = plt.subplots( nrows=1, ncols=2, figsize=(12,6), dpi=150)

exmax = fov_as/2
extent = [-exmax,exmax,-exmax,exmax]

im = ax[0].imshow(monopsf_onax[0].data, norm=LogNorm(), cmap=plt.get_cmap('gist_heat'), extent=extent)
circ_in = Circle((0,0),2.6,edgecolor='cyan', facecolor='none', linewidth=2)
circ_out = Circle((0,0),9,edgecolor='cyan', facecolor='none', linewidth=2)
ax[0].add_patch(circ_in)
ax[0].add_patch(circ_out)
ax[0].set_title('Speckles with Wavefront Maps\n and No Aberrations')
fig.colorbar(im, ax=ax[0], shrink=1) 

im = ax[1].imshow(monopsf_offax[0].data, norm=LogNorm(), cmap=plt.get_cmap('gist_heat'), extent=extent)
circ_in = Circle((0,0),2.6,edgecolor='cyan', facecolor='none', linewidth=2)
circ_out = Circle((0,0),9,edgecolor='cyan', facecolor='none', linewidth=2)
ax[1].add_patch(circ_in)
ax[1].add_patch(circ_out)
ax[1].set_title('Speckles with Wavefront Maps\n and No Aberrations')
fig.colorbar(im, ax=ax[1], shrink=1) 
```

```python
# Form a simple bright companion scene by summing the on-axis and off-axis PSF, the latter scaled by 1E-7.
offax_peak = monopsf_offax[0].data.max()
comb_img = (monopsf_onax[0].data + monopsf_offax[0].data*1e-7)/offax_peak # Contrast units
comb_hdu = fits.PrimaryHDU(comb_img, header=monopsf_onax[0].header)
comb_hdulist = fits.HDUList([comb_hdu])

webbpsf.display_psf(comb_hdulist,vmin=1e-9, vmax=3e-8)
```
